```python
# Import the modules

import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
from sklearn.preprocessing import StandardScaler
import hvplot.pandas
```

```python
# Read CSV File (AllOthers.csv)

csvpath = Path("Projects/03-Project3/AllOthers_1.csv")
```

```python
# Create All Others DataFrame

all_others_df = pd.read_csv(Path("AllOthers_1.csv"), parse_dates=True, index_col="Quarter", infer_datetime_format=True)
```

```python
# Sort Data (Ascending Order)

all_others_df.sort_index(inplace=True)
```

```python
# Review the DataFrame 

all_others_df
```

```python
# Formatting Data (removing commas from Data)

all_others_df["Total_all_other"]=all_others_df["Total_all_other"].str.replace(",","")
all_others_df["Total E_commerce"]=all_others_df["Total E_commerce"].str.replace(",","")
all_others_df["Food_&_Beverage"]=all_others_df["Food_&_Beverage"].str.replace(",","")
all_others_df["Health_&_Personal_Care"]=all_others_df["Health_&_Personal_Care"].str.replace(",","")
all_others_df["Sporting_Goods"]=all_others_df["Sporting_Goods"].str.replace(",","")
all_others_df["Miscellaneous"]=all_others_df["Miscellaneous"].str.replace(",","")

all_others_df
```

```python
# Change dtype

all_others_df = all_others_df.astype("float")
all_others_df
```

```python
# Create DataFrame1

all_others_split_df = all_others_df[["Food_&_Beverage","Health_&_Personal_Care", "Sporting_Goods", "Miscellaneous"]]

all_others_split_df

```

```python
# Plot Df

all_others_split_df.hvplot(title= "E-commerce Sales- in millions (2018-2023)", width=900, height=500)
```
